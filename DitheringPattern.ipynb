```python
from lsst.daf.butler import Butler
import pylab
```

```python
repo_path = '/sdf/data/rubin/repo/ir2/butler.yaml'
collections='LSSTCam/calib'
butler = Butler(repo_path,collections=['u/snyder18/spot_13242/gridfit_run1','u/snyder18/spot_13243/gridfit_run1'])
registry = butler.registry
```

```python
collections=['u/snyder18/spot_13237/gridfit_run1']
refs=list(registry.queryDatasets('gridSpotSrc', collections=collections))
```

```python
x,y=[],[]
for aref in refs:
    src=butler.get(aref, collections=collections)
    x.append(src.getMetadata()['BOTXCAM'])
    y.append(src.getMetadata()['BOTYCAM'])
```

```python
pylab.scatter(x,y)
ax=pylab.gca()
ax.set_aspect('equal')
pylab.xlabel('X in mm')
pylab.ylabel('Y in mm')
```
